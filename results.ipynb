```python
#check if the data normalized or not
epsi = 10**-8
tu = np.mean(nData, axis=0)
tv = np.mean(((tu - nData) ** 2), axis=0)
assert True not in (tu > epsi)
assert True not in (np.abs(tv - 1) > epsi)
```
